```python
#coding: utf-8

import sys
sys.path.insert(0, '/Users/wuchenglin/git/caffe/python')

from matplotlib import pyplot
import caffe
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

caffe_root = '/Users/wuchenglin/git/caffe/'
img_dir = '/Users/wuchenglin/git/imgs_top_10000/imgs/'

# Set the right path to your model file, pretrained model
# and the image you would like to classify.
MODEL_FILE = caffe_root + 'models/bvlc_reference_caffenet/deploy.prototxt'
PRETRAINED = caffe_root + 'models/bvlc_reference_caffenet/bvlc_reference_caffenet.caffemodel'
# IMAGE_FILE = caffe_root + 'examples/images/cat.jpg'
# IMAGE_FILE = img_dir + '1037.32376.0.jpeg'
# IMAGE_FILE = img_dir + '1118.30088.640.jpeg' # water ox
IMAGE_FILE = img_dir + '1990.16732.220.jpeg' # panda
IMAGE_FILE = img_dir + '2010.16542.TB1tCtwGpXXXXbNaXXXWZfhGpXX-140-160.jpg_q60.jpg'
IMAGE_FILES = [img_dir+img for img in [
    '2152.15369.gaoxiao.jpg', # dog
    '1990.16732.220.jpeg', # panda
    '2207.15027.boss-5463098F-9DF173795463098F040100003989E8E4.1.min.1.jpg', #shoe
    '2304.14533.dongwu.jpg', # butterfly
]]

net = caffe.Classifier(MODEL_FILE, PRETRAINED,
                       mean=np.load(caffe_root + 'python/caffe/imagenet/ilsvrc_2012_mean.npy'),
                       channel_swap=(2,1,0),
                       raw_scale=255,
                       image_dims=(256, 256))

print 'Succeed get net and prepare for prediction.'
```

```python
ax = []
fig, (ax) = plt.subplots(1, 4)

input_images = []
for index, i in enumerate(IMAGE_FILES):
    input_image = caffe.io.load_image(i)
    ax[index].imshow(input_image)
    input_images.append(input_image)
    print i
```

```python

prediction = net.predict(input_images)  # predict takes any number of images, and formats them for the Caffe net automatically
for index, p in enumerate(prediction):
    print index
    print 'prediction shape:', p.shape
    plt.plot(p)
    print 'predicted class:', p.argmax()
```
